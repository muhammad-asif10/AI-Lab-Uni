# Concept 1: Lists in Python

```python
# Example 1
robots =["Alpha", "Beta", "Gamma"]

print("Robot List:", robots)
print("First Robot:", robots[0])
print("second  Robot:", robots[1])
print("last Robot:", robots[-1])
```

```python
# Example 2

robots.append("Delta")
robots.remove("Beta")
print("Updated Robot List:", robots)
```

# Concept 2: Dictionaries in Python

```python
robot ={
    "name": "Alpha",
    "bettery":80,
    "status": "Active"
}
print("Robot Name:" , robot["name"])
print("bettery level:", robot["bettery"])
print("Status:", robot["status"])
robot["battery"] = 100              # update a value
robot["task"] = "Line Following"   # add a new key-value pair
print("Updated Robot Info:", robot)
```

# Concept 3: Loops in python

```python
robots=["Alpha","Gamma","Delta"]
for r in robots:
    print("Activating robot:" ,r)
```

```python
battery =5
while battery > 0:
    print("Battery remaining:", battery)
    battery -=1
```

# concept 4: Data Comparison(condition)

```python
battery =65

if battery > 60:
    print("Battery is full _ ready for mission!")
elif battery > 20:
    print("Battery is Medium_ prepare for recharge!")
else:
    print("Battery low _  returning to base")
```

# Mini Al Project: Robot Battery Checker

```python
robots= {
    "alpha":75,
    "Beta": 20,
    "Gamma": 10
}
for name, battery in robots.items():
    if battery > 50:
        print(name, " Battery ok (",battery, "%")
    elif battery <20:
        print(name, " Battery very low (",battery, "%")
    else:
        print(name, "Low Battery  (",battery, "%")
```
